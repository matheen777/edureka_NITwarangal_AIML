```python
import pandas as pd

# [Example] Read datasets
data = pd.read_csv("assign1_SECTION1.csv",header=None)
m_data=pd.DataFrame({'Registration':data[5],'Name':data[6],'Date':data[7],'ClassAttended':data[11]})
unique_reg=m_data.Registration.unique()
unique_date=m_data.Date.unique()

unique_date_list=unique_date.tolist()
unique_date_list.insert(0,'RegistrationNo')
unique_date_list.insert(1,'Name')

#Create a new data DataFrame
df=pd.DataFrame(columns=unique_date_list)

#adding rows into dataframe
indx=0
for i in unique_reg:
    nlist=list()
    nlist.append(i)
    nlist.append(m_data[m_data['Registration']==i]['Name'].head(1).item())
    for j in unique_date:
        s=m_data[m_data['Registration']==i]['Date']
        if s[s.isin([j])].empty:
            summ=0
        else:
            summ=m_data[m_data['Registration']==i].groupby('Date').get_group(j).sum()['ClassAttended']
        nlist.append(summ)
    df.loc[indx]=nlist
    indx=indx+1
print(df)

#reading testcase.txt file
with open('testcase.txt') as f:
    flist=f.read().splitlines()

#first line of the file which tell us how many output files needs to created
nfiles=flist[0]

#list of lines withou first line
flist=flist[1:]
#creating list of tuples to pair the RegistrationNo and date
flist=list(zip(*[iter(flist)]*2))

#writting outputs into n number of files
index=1
for i in flist:
    regno=i[0]
    dd=i[1]
    classattended=df[df['RegistrationNo']==regno][dd].item()
    writefile=open('output'+str(index)+'.txt','w')
    writefile.write(regno+"\n")
    writefile.write(str(classattended))
    writefile.close()
    index=index+1
```
