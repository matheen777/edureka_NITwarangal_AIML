```python
import numpy as np
train=train.drop(['Unnamed: 32','id'],axis=1)


df=pd.DataFrame(train['perimeter_worst'])
df['pw_bins']=pd.cut(x=train['perimeter_worst'],bins=[53,85,125,252])
df['pw_labels']=pd.cut(x=train['perimeter_worst'],bins=[53,85,125,252],labels=['Low','Medium','High'])

df['diagnosis']=train['diagnosis']

print('length of column:',len(df['pw_labels']))
print(df['pw_labels'].value_counts())
print('='*20)
print('medium probability:',df['pw_labels'].value_counts()['Medium']/455)
print('Low probability',df['pw_labels'].value_counts()['Low']/455)
print('high probability',df['pw_labels'].value_counts()['High']/455)
print('*'*100)
print('length column:',len(train['diagnosis']))
print(train['diagnosis'].value_counts())
print('='*20)
print('probability of Benin(safe):',train['diagnosis'].value_counts()['B']/455)
print('probability of malignant(danger):',train['diagnosis'].value_counts()['M']/455)

print('Joint probability distribution:')
print(df.groupby(['diagnosis','pw_labels']).size()/455)

df.head()

cpd=list()
for i,j in zip(df['pw_labels'],df['diagnosis']):
    if (i=='Low' and j=='B'):
        cpd.append(((df.groupby(['diagnosis','pw_labels']).size()/455)['B','Low'])/((df.groupby(['diagnosis','pw_labels']).size()/455)['B'].values.sum()))
    elif (i=='Medium' and j=='B'):
        cpd.append(((df.groupby(['diagnosis','pw_labels']).size()/455)['B','Medium'])/((df.groupby(['diagnosis','pw_labels']).size()/455)['B'].values.sum()))
    elif (i=='High' and j=='B'):
        cpd.append(((df.groupby(['diagnosis','pw_labels']).size()/455)['B','High'])/((df.groupby(['diagnosis','pw_labels']).size()/455)['B'].values.sum()))
    elif (i=='Medium' and j=='M'):
        cpd.append(((df.groupby(['diagnosis','pw_labels']).size()/455)['M','Medium'])/((df.groupby(['diagnosis','pw_labels']).size()/455)['M'].values.sum()))
    elif (i=='High' and j=='M'):
        cpd.append(((df.groupby(['diagnosis','pw_labels']).size()/455)['M','High'])/((df.groupby(['diagnosis','pw_labels']).size()/455)['M'].values.sum()))
    else: None


df['CPD']=np.round(cpd,3)
print(df.head())
```
